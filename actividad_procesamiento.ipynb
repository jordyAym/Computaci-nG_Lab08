```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Leer las imgenes
img1 = cv2.imread("img1.jpg")
img2 = cv2.imread("img2.jpg")
img3 = cv2.imread("img3.jpg")

# Verifica si las imgenes se cargaron
if img1 is None or img2 is None or img3 is None:
    print("Una o ms imgenes no se encontraron.")
else:
    # Obtener las dimensiones mximas (alto, ancho)
    h_max = max(img1.shape[0], img2.shape[0], img3.shape[0])
    w_max = max(img1.shape[1], img2.shape[1], img3.shape[1])
    dim_max = (w_max, h_max)

    # Redimensionar todas las imgenes al tamao de la ms grande
    img1_resized = cv2.resize(img1, dim_max)
    img2_resized = cv2.resize(img2, dim_max)
    img3_resized = cv2.resize(img3, dim_max)

    # Extraer canales individuales
    canal_r = img1_resized[:, :, 2]  # Rojo
    canal_g = img2_resized[:, :, 1]  # Verde
    canal_b = img3_resized[:, :, 0]  # Azul

    # Combinar los canales en una nueva imagen
    combinada = cv2.merge([canal_b, canal_g, canal_r])

    # Mostrar imagen combinada
    plt.imshow(cv2.cvtColor(combinada, cv2.COLOR_BGR2RGB))
    plt.title("Imagen combinada (R-G-B)")
    plt.axis("off")
    plt.show()

    # Guardar imagen combinada
    cv2.imwrite("combinada_rgb.png", combinada)

    # Crear imagen negativa (invertir colores)
    negativa = 255 - combinada
    cv2.imwrite("negativa.png", negativa)

    # Convertir a escala de grises
    gris = cv2.cvtColor(negativa, cv2.COLOR_BGR2GRAY)
    cv2.imwrite("gris.png", gris)

    # Mostrar negativo y gris
    plt.figure(figsize=(10,4))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(negativa, cv2.COLOR_BGR2RGB))
    plt.title("Negativa")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(gris, cmap='gray')
    plt.title("Escala de grises")
    plt.axis("off")
    plt.show()

    print("Imgenes procesadas y guardadas: 'combinada_rgb.png', 'negativa.png', 'gris.png'")
```
